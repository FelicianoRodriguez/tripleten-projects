# Paso 2. Entrenar y probar el modelo para cada región

## 2.1 Dividir los datos en conjuntos de entrenamiento y validación

### Función

```python
# Función para dividir características y objetivo
def split_features_target(df):
    features = df[['f0', 'f1', 'f2']]
    target = df['product']

    return features, target
```

```python
def split_train_valid(features, target):
    features_train, features_valid, target_train, target_valid = train_test_split(
        features, 
        target, 
        test_size=0.25, 
        random_state=12345)
    
    return features_train, features_valid, target_train, target_valid
```

### Región 01

```python
features_reg1, target_reg1 = split_features_target(df_01)
features_train_reg1, features_valid_reg1, target_train_reg1, target_valid_reg1 = split_train_valid(features_reg1, target_reg1)

print(f"Features del conjunto de entrenamiento {features_train_reg1.shape}")
print(f"Features del conjunto de validación {features_valid_reg1.shape}")
print(f"Objetivo del conjunto de entrenamiento {target_train_reg1.shape}")
print(f"Objetivo del conjunto de validación {target_valid_reg1.shape}")
```

### Región 02

```python
features_reg2, target_reg2 = split_features_target(df_02)
features_train_reg2, features_valid_reg2, target_train_reg2, target_valid_reg2 = split_train_valid(features_reg2, target_reg2)

print(f"Features del conjunto de entrenamiento {features_train_reg2.shape}")
print(f"Features del conjunto de validación {features_valid_reg2.shape}")
print(f"Objetivo del conjunto de entrenamiento {target_train_reg2.shape}")
print(f"Objetivo del conjunto de validación {target_valid_reg2.shape}")
```

### Región 03

```python
features_reg3, target_reg3 = split_features_target(df_03)
features_train_reg3, features_valid_reg3, target_train_reg3, target_valid_reg3 = split_train_valid(features_reg3, target_reg3)

print(f"Features del conjunto de entrenamiento {features_train_reg3.shape}")
print(f"Features del conjunto de validación {features_valid_reg3.shape}")
print(f"Objetivo del conjunto de entrenamiento {target_train_reg3.shape}")
print(f"Objetivo del conjunto de validación {target_valid_reg3.shape}")
```

- Se separo las características y los objetivos.
- Finalmente se separo en conjunto de entrenamiento y validación en una propoción de 75:25 respectivamente.
- Esto se realizo para los 3 diferentes datasets.


## 2.2 Entrenar el modelo y predicciones para el conjunto de validación

### Función

```python

# Función para entrenar el modelo
def get_prediction(features_train, target_train, features_valid):
    model = LinearRegression()
    model.fit(features_train, target_train)
    predicted_valid = model.predict(features_valid)

    return predicted_valid
```

### Región 01

```python
# Obtener las predicciones de la región 1
predicted_valid_reg1 = get_prediction(features_train_reg1, target_train_reg1, features_valid_reg1)

print(predicted_valid_reg1)
```

### Región 03

```python
predicted_valid_reg3 = get_prediction(features_train_reg3, target_train_reg3, features_valid_reg3)

print(predicted_valid_reg3)
```

## 2.4 volumen promedio de las reservas previstas y el RECM del modelo.

### Función

```python
# Volumen medio de respuestas predichas
def print_predicted_valid(predicted_valid):
    mean_predicted = predicted_valid.mean()
    print(f"Volumen medio de reservas predicho {mean_predicted:.2f}")

# Volumen medio de respuestas correctas
def print_target_valid(target_valid):
    mean_y_valid = target_valid.mean()
    print(f"Volumne medio de reservas correcto {mean_y_valid:.2f}")

# Imprimir raiz del error cuadratico medio
def print_rmse(target_valid, predicted_valid):
    rmse = mean_squared_error(target_valid, predicted_valid) ** 0.5
    print(f'Raiz del error cuadratico medio {rmse}')
```

### Región 01

```python
print_predicted_valid(predicted_valid_reg1)
print()
print_target_valid(target_valid_reg1)
print()
print_rmse(target_valid_reg1, predicted_valid_reg1)
```

```python
print_predicted_valid(predicted_valid_reg2)
print()
print_target_valid(target_valid_reg2)
print()
print_rmse(target_valid_reg2, predicted_valid_reg2)
```

### Región 03

```python
print_predicted_valid(predicted_valid_reg3)
print()
print_target_valid(target_valid_reg3)
print()
print_rmse(target_valid_reg3, predicted_valid_reg3)
```

# Paso 5. Calcular riesgos y ganancias para cada región

## 5.1 bootstrapping

### Región 1

```python
revenues = []

state = np.random.RandomState(12345)

for i in range(1000):
    subsample = best2_wells_real_reg1.sample(n=500, replace=True, random_state=state)
    subsample = pd.Series(subsample).sort_values(ascending=False).head(200)
    revenues.append(revenue(subsample.sum()))

print(f"Tamaño: {len(revenues)}\n")
print(revenues[:20])
```

### Región 2

```python
revenues_reg2 = []

for i in range(1000):
    subsample = best2_wells_real_reg2.sample(n=500, replace=True, random_state=state)
    subsample = pd.Series(subsample).sort_values(ascending=False).head(200)
    revenues_reg2.append(revenue(subsample.sum()))

print(f"Tamaño: {len(revenues)}\n")
print(revenues_reg2[:20])
```

### Región 3

```python
revenues_reg3 = []

for i in range(1000):
    subsample = best2_wells_real_reg3.sample(n=500, replace=True, random_state=state)
    subsample = pd.Series(subsample).sort_values(ascending=False).head(200)
    revenues_reg3.append(revenue(subsample.sum()))

print(f"Tamaño: {len(revenues)}\n")
print(revenues_reg3[:20])
```

## 5.2 Intervalo de confianza del 95% y el riesgo de pérdida

### Región 1

```python
print(f"Beneficio promedio: {np.mean(revenues):.2f}")
```

```python
# Obtener el intervalo de confianza de 95%
rev = pd.Series(revenues)

lower = rev.quantile(0.025)
upper = rev.quantile(0.075)

print("Invervalo de confianza")
print(f"{lower:.2f}")
print(f"{upper:.2f}")
```

```python
# Calcular el riesgo de pérdida
loss = sum(revenue < 0 for revenue in revenues)
rev_len = len(rev)
risk_loss = (loss / rev_len) * 100
print(f"Riesgo de pérdida {risk_loss}%")
```

### Región 2

```python
print(f"Beneficio promedio: {np.mean(revenues_reg2):.2f}")
```

```python
# Obtener el intervalo de confianza de 95%
rev_reg2 = pd.Series(revenues_reg2)

lower_reg2 = rev_reg2.quantile(0.025)
upper_reg2 = rev_reg2.quantile(0.075)

print("Intervalo de confianza")
print(f"{lower_reg2:.2f}")
print(f"{upper_reg2:.2f}")
```

```python
# Calcular el riesgo de pérdida
loss_reg2 = sum(revenue < 0 for revenue in revenues_reg2)
risk_loss_reg2 = (loss_reg2 / len(revenues_reg2)) * 100
print(f"Riesgo de pérdida {risk_loss_reg2:.2f}%")
```

### Región 3

```python
print(f"Beneficio promedio: {np.mean(revenues_reg3):.2f}")
```

```python
# Obtener el intervalo de confianza de 95%
rev_reg3 = pd.Series(revenues_reg3)

lower_reg3 = rev_reg3.quantile(0.025)
upper_reg3 = rev_reg3.quantile(0.075)

print("Intervalo de confianza")
print(f"{lower_reg3:.2f}")
print(f"{upper_reg3:.2f}")
```

```python
# Calcular el riesgo de pérdida
loss_reg3 = sum(r < 0 for r in revenues_reg3)
risk_loss_reg3 = (loss_reg3 / len(revenues_reg3)) * 100
print(f"Riesgo de pérdida {risk_loss_reg3:.2f}%")
```
