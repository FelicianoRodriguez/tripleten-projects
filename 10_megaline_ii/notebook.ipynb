Desarrolla un modelo con la mayor exactitud posible. En este proyecto, el umbral de exactitud es 0.75. Usa el dataset para comprobar la exactitud.

```python
# Importar las librerías
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

## Paso 1. Abre y examina el archivo de datos

```python
# Leer el dataset
df = pd.read_csv('datasets/users_behavior.csv')
```

```python
print(df.info())
```

```python
print(df.head())
```

## Paso 2. Segmenta los datos fuente en un conjunto de entrenamiento, uno de validación y uno de prueba.

```python
# Segmentar los datos
train, other = train_test_split(df, test_size=0.40, random_state=54321)
test, valid = train_test_split(other, test_size=0.50, random_state=54321)
```

```python
print('Total: ', df.size)
print('Conjunto de entrenamiento', train.size)
print('Conjunto de prueba', test.size)
print('Conjunto de validación', valid.size)
```

```python
X_train = train.drop('is_ultra', axis=1)
y_train = train['is_ultra']
X_valid = valid.drop('is_ultra', axis=1)
y_valid = valid['is_ultra']
X_test = test.drop('is_ultra', axis=1)
y_test = test['is_ultra']
```

Cada conjunto de datos se repartio de la siguiente manera:
- Entrenamiento: 60%
- Validación: 20%
- Prueba: 20%

## Paso 3. Investiga la calidad de diferentes modelos cambiando los hiperparámetros.

Modelos a investigar:
- árbol de decisión
- bosque aleatorio
- regresión logística

### Árbol de decisión

```python
for depth in range(1, 11):
    model_tree = DecisionTreeClassifier(random_state=12345, max_depth=depth)
    model_tree.fit(X_train, y_train)
    predic_tree = model_tree.predict(X_valid)
    score_train = model_tree.score(X_train, y_train)
    score_valid = model_tree.score(X_valid, y_valid)

    print(f'Conjunto de entrenamiento - Depth: {depth} - Accuracy: {score_train}')
    print(f'Conjunto de validación - Depth: {depth} - Accuracy: {score_valid}')
```

### Bosque aleatorio

```python
best_score = 0
best_est = 0
best_depth = 0

for est in range(10, 51, 10):
    for depth in range(1, 11):
        model_forest = RandomForestClassifier(random_state=12345, n_estimators=est, max_depth=depth)
        model_forest.fit(X_train, y_train)
        score_valid = model_forest.score(X_valid, y_valid)
        score_train = model_forest.score(X_train, y_train)

        print('Conjunto de entrenamiento (n_estimators = {}, depth = {}): {})'.format(est, depth, score_train))
        print('Conjunto de validación (n_estimators = {}, depth = {}): {})'.format(est, depth, score_valid))


        if score_valid > best_score:
            best_score = score_valid
            best_est = est
            best_depth = depth

print()
print('La mejor exactitud en el conjunto de validación (n_estimators = {}, depth = {}): {})'.format(best_est, best_depth, best_score))
```

### Regresión logística

```python
model_logistic = LogisticRegression(random_state=12345, solver='liblinear')
model_logistic.fit(X_train, y_train)
score_train = model_logistic.score(X_train, y_train)
score_valid = model_logistic.score(X_valid, y_valid)

print('Accuracy del modelo de regresión logística en el conjunto de entrenamiento: ', score_train)
print('Accuracy del modelo de regresión logística en el conjunto de validación: ', score_valid)
```

- Se realizó el análisis de los modelos: árbol de decisión, bosque aleatorio, regresión logística, cambiando sus diferentes hiperparámetros. 
- Con la ayuda de bucles se probaron los diferentes hiperparámetros de los 3 diferentes modelos.
- Se imprimió el accuracy con el conjunto de validación y el conjunto de entrenamiento para asegurarnos de que no exista overfitting o underfitting.
- Se encontro la mejor exactitud en el modelo bosque aleatorio con n_estimators = 20 y depth = 5, resultando en el conjunto de validación la puntuación de: 0.8133748055987559.

## Paso 4. Comprueba la calidad del modelo usando el conjunto de prueba

```python
model_forest = RandomForestClassifier(random_state=12345, n_estimators=20, max_depth=5)
model_forest.fit(X_train, y_train)
score = model_forest.score(X_test, y_test)

print('Conjunto de prueba (n_estimators = {}, depth = {}): {})'.format(20, 5, score))
```

La puntuación máxima de exatitud se alcanzo con el modelo bosque aleatorio con los hiperparámetros n_estimators=20 y depth=5, superando el umbral de 0.75. Dando una exactitud en el conjunto de prueba de 0.81.
